```python
try:
    import urllib2 as ulib
except ImportError:
    import urllib3 as ulib

try:
    from StringIO import BytesIO as io
except ImportError:
    from io import BytesIO as io

try:
    from urllib import urlencode as urlencode
except ImportError:
    from urllib.parse import urlencode as urlencode
    
try:
    from urllib import urlopen as urlopen
except ImportError:
    from urllib.request import urlopen as urlopen
    
try:
    from urllib2 import HTTPError as HTTPError
except ImportError:
    from urllib.error import HTTPError as HTTPError

import ast
import pandas as pd
```

```python
API_URL = 'https://fb55.carto.com:443/api/v2/sql?q='

def queryCartoDB(query, formatting = 'CSV', source = API_URL):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    
    data = urlencode({'format': formatting, 'q': query}).encode("utf-8")
    try:
        response = urlopen(source, data)
        return response.read()
    except HTTPError as e:
        raise (ValueError('\n'.join(ast.literal_eval(e.readline())['error'])))
        
def get_data(query):
    try:
        return pd.read_csv(io(queryCartoDB(query)), sep = ',')
    except ValueError as v:
        print (str(v))
```

```python
query = '''SELECT * FROM fb55.citibike;'''
table = get_data(query)
table
```

## Task 1 - Familiarize with SQL Clauses

```python
# sort data by start_station_id, tripduration; only checking trips with duration <= 3 hours 
query = '''
SELECT * FROM fb55.citibike
WHERE tripduration <= 10800
ORDER BY start_station_id ASC, tripduration DESC;
'''

table = get_data(query)
table.head(5)
```

```python
# Only show the top 10 records
query = '''
SELECT * FROM fb55.citibike
ORDER BY cartodb_id ASC
LIMIT 10;
'''

table = get_data(query)
table.head(10)
```

```python
# Only show the last 10 records
query = '''
SELECT q.* FROM 
    (SELECT * FROM fb55.citibike
    ORDER BY cartodb_id DESC
    LIMIT 10) q
ORDER BY q.cartodb_id ASC
'''

table = get_data(query)
table.head(10)
```

```python
# List all unique start_station_id values
query = '''
SELECT DISTINCT start_station_id FROM fb55.citibike;
'''

table = get_data(query)
table.head(5)
```

```python
#Count the number of trips
query = '''
SELECT COUNT(cartodb_id)FROM fb55.citibike;
'''

table = get_data(query)
table.head(5)
```

```python
# Find the average/min/max trip duration
query = '''
SELECT AVG(tripduration), MIN(tripduration), MAX(tripduration)  FROM fb55.citibike;
'''

table = get_data(query)
table.head(5)
```

## Task 2 - Working with date/time

```python
# Selecting trips started on Feb-02-2015 only 
query = '''
SELECT * FROM fb55.citibike 
WHERE EXTRACT(MONTH FROM starttime) = 2 AND
EXTRACT(DAY FROM starttime) = 2 AND
EXTRACT(YEAR FROM starttime) = 2015
'''

table = get_data(query)
table.head(5)
```

```python
# Selecting trips started on the weekends; What are average trip duration during weekends? 
query = '''
SELECT AVG(tripduration) FROM fb55.citibike 
WHERE EXTRACT(DOW from starttime) IN (0,6)
'''

table = get_data(query)
table.head(5)
```

```python
# Same for weekday
query = '''
SELECT AVG(tripduration) FROM fb55.citibike 
WHERE EXTRACT(DOW from starttime) IN (1,2,3,4,5) 
'''

table = get_data(query)
table.head(5)
```

## Task 3 - Working with Space

```python
# Showing the list of start station locations 
query = '''
SELECT start_station_id, start_station_name FROM fb55.citibike 
GROUP BY  start_station_id, start_station_name
'''

table = get_data(query)
table.head(5)
```

```python
# Showing the number of trips started per station 
query = '''
SELECT start_station_id, start_station_name, count(start_station_id) as Count FROM fb55.citibike 
GROUP BY  start_station_id, start_station_name
'''

table = get_data(query)
table.head(5)
```

```python
# only for stations within 500m of Time Square

# taking one lat/long as roughly 111 km as given here:
# https://gis.stackexchange.com/questions/142326/calculating-longitude-length-in-miles

# 111000/500 = 222
query = ''' 
SELECT start_station_id, start_station_name, COUNT(start_station_id) as Count from fb55.citibike
WHERE (start_station_longitude - -73.9857) > .000222 
AND (start_station_latitude - 40.7577) > .000222
GROUP BY start_station_id ,start_station_name
'''

table= get_data(query)
table.head(5)
```

## Task 4 - Putting it all together

```python
# Find the station that had the longest average trip duration during weekends and within 500m of TimeSquare
query = '''

SELECT tripduration, start_station_id, start_station_name from fb55.citibike
WHERE (start_station_longitude - -73.9857) > .000222 
AND (start_station_latitude - 40.7577) > .000222
AND EXTRACT(DOW from starttime) IN (0,6)
GROUP BY tripduration, start_station_id ,start_station_name
ORDER BY tripduration DESC
LIMIT 1

'''

table = get_data(query)
table.head(5)
```

```python
#create lines for trips started from stations within 500m of Times Squares and lasted less than 2 hours.
query = '''

SELECT tripduration, start_station_id, start_station_name, count(*) from fb55.citibike
WHERE (start_station_longitude - -73.9857) > .000222 
AND (start_station_latitude - 40.7577) > .000222
AND tripduration < 7200
GROUP BY tripduration, start_station_id ,start_station_name


'''

table = get_data(query)
table.head(5)
```
