```python
import fix_yahoo_finance as yf
import matplotlib.pyplot as plt
%pylab inline
```

```python
# Grab Apple share data for 10 years, 2006 - 2016

start, end = "2006-01-01", "2016-12-31"

aapl_all = yf.download('aapl', start=start, end=end)
```

```python
aapl_all.head()
```

```python
aapl = aapl_all['Adj Close']
```

```python
plt.figure(figsize = (12, 10))
aapl.plot(color = 'blue')
plt.ylabel('$/Share', fontsize = 15)
plt.xlabel('Date', fontsize = 15)
plt.title('Apple Shares 2006 - 2016', fontsize = 20)
```

The above plot shows the dollars per share for Apple for a period of 10 years (2006 - 2016). The various highs and lows can be attributed to real world events (for example the recession around 2008 - 2009).
